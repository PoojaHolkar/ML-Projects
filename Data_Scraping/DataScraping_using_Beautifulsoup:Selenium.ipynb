# Web Scraping Indeed.com

```python
#importing libraries
import sys
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
```

```python
#Applying Beautiful Soup to scrape the indeed jobs for Job title “Product Manager” ,where location is “San Francisco, CA”
page=requests.get("https://www.glassdoor.co.in/Reviews/IBM-Reviews-E354.htm").text
Soupobject=BeautifulSoup(page,'lxml')
```

```python
Soupobject
```

```python
#Pagenation-picking links for next pages
cookedurl=[]
for i in range(0,300,10):
        link = "https://www.indeed.com/jobs?q=Product+Manager&l=San+Francisco%2C+CA" + str(i)
        cookedurl.append(link)
print(cookedurl)
    
```

```python
# creating a function to pick all column values from the parsed data
jobs=[]
locations=[]
summary=[]

def extractData(link):
    page=requests.get(link).text
    for JT in Soupobject.findAll("a",{"class": "jobtitle turnstileLink" }):
        jobs.append(JT.text.strip())
    spans = Soupobject.findAll('span', attrs={'class': 'location'})
    for span in spans:
        locations.append(span.text)
    for sum in Soupobject.findAll('div', {'class': 'summary'}):
        summary.append(sum.text.strip())
```

```python
#Extracting data for all pages 
for link in cookedurl:
    extractData(link)
```

```python
print(jobs)
print(locations)
print(summary)
```

```python
#creating dataframe for all the indeed jobs title ,locations and summary detail
PMscraping=pd.DataFrame({'Job Title':jobs,'Location':locations,'Job_Description':summary})
PMscraping.to_csv("/Users/poojaholkar/ISB/RESIDENCY_1/PMscraping.csv",sep=",",index=False)
```

#  Identify Key Tech Hires in the Hyderabad area

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support import expected_conditions as EC
import os
import csv
import pandas as pd
from math import ceil
```

```python
locationOfWebdriver = "/usr/local/Caskroom/chromedriver/77.0.3865.40/chromedriver"
driver = webdriver.Chrome(locationOfWebdriver)
```

```python
url = 'https://www.naukri.com/information-technology-jobs'
driver.get(url)
elem_1=driver.find_element_by_name("ql")
elem_1.clear()
elem_1.send_keys("Hyderabad/Secunderabad")
elem_1.send_keys(Keys.RETURN)
baseurl = driver.current_url
```

```python
print(baseurl)
```

```python
#Pagenation-picking links for next pages
cookurl=[]
for i in range(1,23,1):
        link = baseurl+"-"+str(i)
        cookurl.append(link)

print(cookurl)
```

```python
#Applying Beautiful Soup to scrape the naukri jobs for Job title “Information technolgy” ,where location is “Hyderabad”
import sys
import requests
from bs4 import BeautifulSoup
page=driver.page_source
soupobject=BeautifulSoup(page,'lxml')
soupobject
```

```python
company=[]
segment=[]
skills=[]

def hydjobsdata(link):
    page=requests.get(link).text
    for com in soupobject.findAll('span', attrs={'class': 'org'}):
        company.append(com.text.strip())
    for seg in soupobject.findAll('li', attrs={'class': 'desig'}):
        segment.append('Information Tech/Information Services')
    for skill in soupobject.findAll('span', {'class': 'skill'}):
        skills.append(skill.text.strip())
        
```

```python
#Extracting data for all pages 
for link in cookurl:
    hydjobsdata(link)
```

```python
print(company)
print(segment)
print(skills)
```

```python
#creating dataframe to group top companies
df1=pd.DataFrame({'Company_Name':company,'Segment':segment,'Job_Skills':skills})
```

```python
#counting number of positions for one company at a time
group=df1.groupby(['Company_Name','Segment','Job_Skills'])
df2=group['Company_Name'].count().sort_values(ascending=False).reset_index(name='Num_of_positions')
df2
```

```python
finaldf = df2[['Company_Name','Num_of_positions','Segment','Job_Skills']]
finaldf
```

```python
#creating csv file from the dataframe
finaldf.to_csv("/Users/poojaholkar/ISB/RESIDENCY_1/hyderabad.csv",sep=",",index=False)
```
